```python
import pandas as pd
data=pd.read_csv("dataset.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
def rankweekendplaces(cityname):
    cityplaces = data[data["City"].str.lower() == cityname.lower()]
    
    cityplaces = cityplaces.sort_values(by=["Google review rating", "Number of google review in lakhs"], ascending=[False, False])
    topplaces = cityplaces[["Name", "Type", "Google review rating", "Number of google review in lakhs", "Best Time to visit"]]

    return topplaces.head(10)
```

```python
city = input("Enter city name: ")
topweekendplaces = rankweekendplaces(city)
print("\nTop Weekend Places in", city)
print(topweekendplaces)
```
